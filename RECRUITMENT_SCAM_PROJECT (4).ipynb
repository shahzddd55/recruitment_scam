# Dimentionality reduction

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA # Import PCA

# Ensure 'df' is a DataFrame and the columns exist in the DataFrame
columns_to_scale = ['title', 'location', 'department', 'salary_range', 'company_profile',
                    'description', 'requirements', 'benefits', 'telecommuting',
                    'has_company_logo', 'has_questions', 'employment_type',
                    'required_experience', 'required_education', 'function',
                    'fraudulent', 'in_balanced_dataset']

# Create a LabelEncoder instance
encoder = LabelEncoder()

# Encode categorical features in the columns to be scaled
# Check if column exists before encoding
for column in columns_to_scale:
    if column in df.columns and df[column].dtype == 'object':
        df[column] = encoder.fit_transform(df[column])

# Standardize the data after encoding
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[columns_to_scale])

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Add the principal components back to the DataFrame
df['pca'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]
```
